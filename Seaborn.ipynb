# Distribution Plots

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#distplot allows the distribution of unovariate set of obsevations,  Histogram
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=50)
#KDE:Hernal Density Estimation plot
```

```python
#jointplot allows to combine two distributions,# essentially combine two diffrenet distribution plots by two variable

sns.jointplot(x='total_bill',y='tip',data=tips ,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips ,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips ,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips ,kind='kde')
```

```python
#pairplot essentially going to plot pairwise relationships across an entire dataframe atleast for the numerical columns.

sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
#palette is used for colors
```

```python
sns.rugplot(tips['total_bill'])
#it just draw a dash mark for every points on the uniform for unique variant distribution essentially one single variable
```

# Categorical Plots

```python
#bar plot allows to aggregate the categorical data based on some function., this is a visualization of a group by action.
sns.barplot(x='sex', y='total_bill',data=tips)
```

```python
import numpy as np
sns.barplot(x='sex', y='total_bill',data=tips,estimator=np.std)
```

```python
#countplot explicitily count the number of occurrences
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips , hue='smoker')
```

```python
sns.stripplot(x='day',y='total_bill', data=tips)
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=True, hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=True, hue='sex', split=True)
```

```python
sns.factorplot(x='day',y='total_bill', data=tips, kind='bar')
```

# Matrix Plot

```python
tips_new = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips_new.head()
```

```python
flights.head()
```

```python
#heat map is the primary way of showing some sort of matrix plot .In order to work properly the data should be in a matrix form
#ie, index name and column name match up
tcr = tips_new.corr()
```

```python
sns.heatmap(tcr ,annot=True)
```

```python
flights
```

```python
fp = flights.pivot_table(values='passengers',index='month',columns='year')
fp
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp, linecolor = 'white' , linewidth = 1)
```

```python
#The clustermap uses hierarchal clustering to produce a clustered version of the heatmap
sns.clustermap(fp)
```

```python
sns.clustermap(fp, standard_scale=1 ,linecolor = 'white' , linewidth = 1)
```
